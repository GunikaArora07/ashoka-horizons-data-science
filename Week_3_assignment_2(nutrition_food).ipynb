```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('nutrition-food.csv', encoding='UTF-16')
```

```python
df.head(5)
```

```python
df.tail()
```

```python
df.fillna(0)
```

```python
df.dtypes
```

```python
exclude_column= 'Unnamed: 0'
for column in df.columns:
    if column != exclude_column:
               df[column] = df[column].replace('-', 0).astype(float)
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
```

```python
plt.plot(df[' Calories'], df[' Fat (g)'], marker= 'o', markersize= '5.0', markeredgecolor= 'black', markerfacecolor= 'blue', color= 'green')
plt.title('Calories vs Fat')
plt.xlabel('Calories')
plt.ylabel('Fat')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(x=df[' Fat (g)'].unique(), height=df[' Fat (g)'].value_counts().to_list(), color='Brown')
plt.xlabel('Fats')
plt.ylabel('Count')
plt.title('Count of Fats', fontsize=20, color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df[' Calories'], df[' Fat (g)'], marker= 'o', markersize= '5.0', markeredgecolor= 'black', markerfacecolor= 'blue', color= 'green')
plt.xlabel('Calories')
plt.ylabel('Fats')
plt.title('Calories vs Fat')

plt.subplot(1,2,2)
plt.bar(x=df[' Fat (g)'].unique(), height=df[' Fat (g)'].value_counts().to_list(), color='Brown')
plt.xlabel('Fats')
plt.ylabel('Count')
plt.title('Count of Fats', fontsize=20, color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df[' Calories'], df[' Fat (g)'], marker= 'o', markersize= '5.0', markeredgecolor= 'black', markerfacecolor= 'blue', color= 'black')
plt.fill_between(df[' Calories'], df[' Fat (g)'], color='Lightblue')
plt.title('Calories vs Fat')
plt.xlabel('Calories')
plt.ylabel('Fat')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(df[[' Calories', ' Fiber (g)']], color='yellow')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=20, color='Black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=' Protein (g)', data=df.reset_index())
plt.xlabel('Protein')
plt.ylabel('Count')
plt.title('Protein Count', fontsize=20, color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(df[[' Calories', ' Fat (g)']], edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=20, color='Red')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```

```python
df_plot = df.drop('Unnamed: 0', axis=1)
df_plot
```

```python
plt.figure(figsize=(10,5))
link = linkage(df_plot, 'ward')
dendrogram(link)
plt.show()
```

```python
plt.figure(figsize=(10,5))
link = linkage(df_plot, 'ward')
dendrogram(link, color_threshold=5, truncate_mode='lastp', p=10)
plt.xticks(rotation=90)
plt.title('Dendrogram Plot', fontsize=20, color='red')
plt.show()
```
